```python
#mutation
 #substitution mutation
 #insertion mutation
 #deletion mutation

dna='ATGCCGTAGCTAAGGCTT'

substitution = dna[:4] + "A" + dna[5:]
print("Substitution:", substitution)

insertion = dna[:5] + "T" + dna[5:]
print("Insertion:", insertion)

deletion = dna[:4] + dna[5:]
print("Deletion:", deletion)
```

```python
#translate all 4 sequence in to protein
dna1=dna
dna2=substitution
dna3=insertion
dna4=deletion
```

```python
def complementary_strand(dna_seq):
    complement_map=str.maketrans("ATCG","TAGC")
    complement=dna_seq.translate(complement_map)
    return complement

comp1=complementary_strand(dna1)
comp2=complementary_strand(dna2)
comp3=complementary_strand(dna3)
comp4=complementary_strand(dna4)

print("Original DNA_sequences:",dna1,dna2,dna3,dna4)
print("Complementary DNA_sequences:",comp1,comp2,comp3,comp4)
```

```python
def dna_to_mrna(complement_seq):
    mrna=complement_seq.replace("T","U")
    return mrna

mrna1=dna_to_mrna(comp1)
mrna2=dna_to_mrna(comp2)
mrna3=dna_to_mrna(comp3)
mrna4=dna_to_mrna(comp4)

print("mRNA_sequences:",mrna1,mrna2,mrna3,mrna4)
```

```python
def split_into_codons(mrna_seq):
    codons=[mrna_seq[i:i+3] for i in
        range(0,len(mrna_seq),3)
        if len(mrna_seq[i:i+3])==3]
    return codons

codons1=split_into_codons(mrna1)
codons2=split_into_codons(mrna2)
codons3=split_into_codons(mrna3)
codons4=split_into_codons(mrna4)

print("Codons:",codons1,codons2,codons3,codons4)
```

```python
codon_table = {
    "UUU": "F", "UUC": "F",
    "UUA": "L", "UUG": "L", "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "AUU": "I", "AUC": "I", "AUA": "I",
    "AUG": "M",
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S", "AGU": "S", "AGC": "S",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "UAU": "Y", "UAC": "Y",
    "CAU": "H", "CAC": "H",
    "CAA": "Q", "CAG": "Q",
    "AAU": "N", "AAC": "N",
    "AAA": "K", "AAG": "K",
    "GAU": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "UGU": "C", "UGC": "C",
    "UGG": "W",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G",
    "UAA": "*", "UAG": "*", "UGA": "*"
}

def trancelate_to_amino_acids(codons):
    amino_acid="".join(codon_table.get(codon,"?")
                      for codon in codons)
    return amino_acid

amino_acid_seq_1=trancelate_to_amino_acids(codons1)
print("Amino acid sequence_1:",amino_acid_seq_1)
amino_acid_seq_2=trancelate_to_amino_acids(codons2)
print("Amino acid sequence_2:",amino_acid_seq_2)
amino_acid_seq_3=trancelate_to_amino_acids(codons3)
print("Amino acid sequence_3:",amino_acid_seq_3)
amino_acid_seq_4=trancelate_to_amino_acids(codons4)
print("Amino acid sequence_4:",amino_acid_seq_4)
```

```python
#find the length of all sequences
proteins = [amino_acid_seq_1,amino_acid_seq_2,amino_acid_seq_3,amino_acid_seq_4]
for i, seq in enumerate(proteins, start=1):
    print(f"Protein {i} length: {len(seq)}")
```

```python
#Compare the original Dna identity presentage with mutated sequences
def sequence_identity(seq1, seq2):
    length = min(len(seq1), len(seq2))
    matches = sum(1 for a, b in zip(seq1[:length], seq2[:length]) if a == b)
    identity=(matches / length) * 100, matches, length
    return identity


mutations = {
    "Substitution": substitution,
    "Insertion": insertion,
    "Deletion": deletion
}

print("Originat DNA:",dna,"\n")
for name, mut_seq in mutations.items():
    identity, matches, length = sequence_identity(dna, mut_seq)
    print(f"{name}: {mut_seq}")
    print(f"  Matches: {matches}/{length}")
    print(f"  Identity: {identity:.2f}%\n")

```
